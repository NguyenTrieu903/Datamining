```python
# I read the dataset with the pandas library.
import pandas as pd
data = pd.read_csv("gender_classification_v7.csv")
```

```python
# information of data
data.info()
```

```python
# Data of values(Max, min, std...)
data.describe()
```

```python
# Data Correlation: Relationship between columns
data.corr()
```

```python
# Naive Bayes Classification
# Change gender type
# Male : 1
# Female : 0
data.gender = [1 if i == "Male" else 0 for i in data.gender]
```

```python
data
```

```python
# x_data
x_data = data.drop(["gender"],axis = 1)

# y_data
y_data = data.gender.values
x_data
```

```python
y_data
```

```python
# Train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size = 0.3, random_state=1)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, f1_score

gnb = GaussianNB()

gnb.fit(x_train, y_train)
y_pred = gnb.predict(x_test)
f1_score_my = f1_score(y_test, y_pred,average='macro')*100
print('f1_score of our model is equal ' + str(round(f1_score_my, 2)) + ' %.')
```
